```python
#Exercise 2.3
#Feito com ajuda do chat gpt

# Função para deslocar uma letra para trás no alfabeto
def caesar_decrypt(text, shift):
    decrypted_text = ""
    for char in text:
        if char.isalpha():  # Apenas letras são deslocadas
            is_upper = char.isupper()
            base = ord('A') if is_upper else ord('a')
            # Aplica o deslocamento ao contrário
            decrypted_char = chr((ord(char) - base - shift) % 26 + base)
            decrypted_text += decrypted_char
        else:
            decrypted_text += char  # Espaços e outros caracteres não são alterados
    return decrypted_text

# Função para aplicar o processo de decifragem 3 vezes
def decrypt_three_times(cipher_text, shift, steps=3):
    decrypted_text = cipher_text
    for _ in range(steps):
        decrypted_text = caesar_decrypt(decrypted_text, shift)
    return decrypted_text

# Texto cifrado e deslocamento (shift)
cipher_text = "OCZ XJMMZXO VINRZM"
shift = 7

# Decifrar o texto
decrypted_message = decrypt_three_times(cipher_text, shift)
print("Texto decifrado:", decrypted_message)
```

```python
#Exercise 3.4
#Feito com ajuda do chatgpt
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# A matriz de A até H
data = np.array([
    [1, 6, 1, 1, 1, 3, 1, 1],
    [1, 4, 4, 6, 1, 5, 3, 7],
    [1, 4, 4, 6, 1, 5, 7, 8],
    [1, 4, 6, 8, 1, 7, 8, 4],
    [1, 5, 7, 7, 2, 0, 0, 5],
    [2, 2, 5, 7, 2, 9, 9, 5],
    [3, 6, 7, 0, 3, 7, 6, 4],
    [4, 6, 9, 5, 3, 3, 3, 7],
    [4, 4, 8, 4, 4, 2, 3, 5],
    [3, 6, 7, 3, 5, 0, 3, 5],
    [5, 4, 7, 2, 5, 8, 7, 5],
    [6, 6, 9, 4, 6, 8, 7, 4],
    [7, 4, 7, 5, 7, 4, 5, 7],
    [8, 6, 0, 2, 7, 2, 4, 3],
    [9, 7, 3, 9, 8, 3, 5, 6]
])

# Calcular a correlação da matriz
corr_matrix = np.corrcoef(data.T)

# Criar o grafo
G = nx.Graph()

# Add nós
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
G.add_nodes_from(nodes)

# Add edges based on positive correlations (ignoring self-loops)
for i in range(len(nodes)):
    for j in range(i + 1, len(nodes)):  # Avoid self-loops and duplicates
        if corr_matrix[i, j] > 0:  # Only positive correlations
            G.add_edge(nodes[i], nodes[j], weight=corr_matrix[i, j])

# Draw the graph
plt.figure(figsize=(8, 8))
pos = nx.spring_layout(G)
edges = G.edges(data=True)

# Draw nodes and edges
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold')

# Draw edges with varying thickness based on correlation weights
weights = [edge[2]['weight'] * 5 for edge in edges]  # scale weights
nx.draw_networkx_edges(G, pos, width=weights)

plt.title("Correlation Network with Positive Weights")
plt.show()
```

```python
#Exercise 4.4
#Feito com auxílio do chatgpt

# Dados da rede dinâmica: [nó_origem, nó_destino, snapshot_inicial, snapshot_final]
edges = [
    (1, 2, 1, 4), (1, 2, 7, 9), (1, 3, 3, 7), (1, 4, 2, 4), (1, 5, 1, 9),
    (2, 3, 1, 2), (2, 4, 1, 3), (2, 4, 6, 7), (2, 4, 9, 9), (3, 5, 5, 7),
    (3, 6, 1, 3), (3, 6, 8, 9), (4, 7, 1, 9), (5, 8, 1, 2), (5, 8, 4, 7),
    (5, 8, 9, 9), (5, 9, 1, 5), (5, 10, 8, 9), (5, 11, 7, 8), (6, 12, 4, 7),
    (6, 13, 2, 3), (7, 12, 2, 6)
]

# Função para agregar as arestas por janelas de tempo
def aggregate_edges(edges, window_size=3, total_snapshots=9):
    # Inicializa as janelas
    num_windows = total_snapshots // window_size
    windows = {i: set() for i in range(1, num_windows + 1)}

    # Processa cada aresta
    for u, v, start, end in edges:
        # Determina as janelas em que essa aresta está presente
        for snapshot in range(start, end + 1):
            window_num = (snapshot - 1) // window_size + 1
            if window_num in windows:
                windows[window_num].add((u, v))

    return windows

# Agrega as arestas
windows = aggregate_edges(edges)

# Exibe o resultado
for window_num, edges_in_window in windows.items():
    print(f"Janela {window_num} (Snapshots {1 + (window_num - 1) * 3}-{window_num * 3}):")
    for edge in edges_in_window:
        print(edge)
    print()
```
